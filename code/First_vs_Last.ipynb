```python
import pandas as pd
import matplotlib.pyplot as plt 
import scipy.stats as st
from scipy.stats import linregress
from scipy.stats import ttest_ind
import os
```

```python
df_first=pd.read_csv('Projects/Project_2/first_season_totals.csv')
df_first.head()
```

```python
df_fifteenth=pd.read_csv('Projects/Project_2/fifteenth_season_totals.csv')
df_fifteenth.head()
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(5)


axes = plt.gca()
axes.xaxis.label.set_size(10)
axes.yaxis.label.set_size(10)

plt.bar(df_first['Player'], df_first['PTS'], color=["red", "purple"])

plt.xlabel("Players")
plt.ylabel("Points for Year")
plt.title("Total Points Score per Player for First Season", fontsize = 30)

plt.savefig('totalscores_first.png')
plt.show()
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(5)


axes = plt.gca()
axes.xaxis.label.set_size(10)
axes.yaxis.label.set_size(10)

plt.bar(df_fifteenth['Player'], df_fifteenth['PTS'], color=["red", "purple"])

plt.xlabel("Players")
plt.ylabel("Points for Year")
plt.title("Total Points Score per Player for Fifteenth Season", fontsize = 30)

plt.savefig('totalscores_fifteenth.png')
plt.show()
```
